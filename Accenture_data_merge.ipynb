### Reaction Table 

```python
import pandas as pd
reaction=pd.read_csv("/content/Reactions.csv")
reaction.head()
```

```python
reaction.info()
```

```python
reaction.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
reaction.drop('Unnamed: 5',axis=1,inplace=True)
```

### Merge Table 

```python
data = content.merge(reaction, how='left', on='User ID')
```

```python
data.head()
```

```python
data = data.merge(reaction_type, how='left', on='Type')
```

```python
data.head()
```

```python
data.info()
```

```python
data.to_csv('combined.csv', index = False)
```

```python
data.groupby('Category').agg({"Type" : "count"}).reset_index().sort_values(by='Type', ascending=False).head(5)
```

```python
data.groupby('Category').agg({"Score" : "sum"}).reset_index().sort_values(by='Score', ascending=False).head()
```
